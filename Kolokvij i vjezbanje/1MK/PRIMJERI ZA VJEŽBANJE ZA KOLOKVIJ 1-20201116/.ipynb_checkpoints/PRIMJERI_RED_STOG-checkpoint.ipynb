## Implementacija reda pomoću DLL


Operacije nad redom:

1. enqueue()     : dodaje na kraj reda.
2. dequeue()     : skida i vrača prvi elemnt u redu
3. first()       : vrača prvi elemnt u redu 
4. size()        : vrača broj elemenata u redu
5. isEmpty()     : provjera da li je red prazan
6. printqueue()  : Ispis svih elemenata reda


```python
# Red implemntiran s DDL
 
# Čvor klasa
class Node: 
   
# Fkonstruktor
    def __init__(self, data): 
        self.data = data # Assign data 
        self.next = None # Initialize next as null 
        self.prev = None # Initialize prev as null 
           
           
# Klasa Queue koja sadrži čvorove Node
class Queue: 
   
    # konstruktor
    def __init__(self): 
        self.head = None
        self.last=None
           
   
#ubacivanje u red
    def enqueue(self, data): 
        if self.last is None: 
            self.head =Node(data) 
            self.last =self.head 
        else: 
            self.last.next = Node(data) 
            self.last.next.prev=self.last 
            self.last = self.last.next
               
               
               
# izbacivanje iz reda
    def dequeue(self): 
   
        if self.head is None: 
            return None
        else: 
            temp= self.head.data 
            self.head = self.head.next
            self.head.prev=None
            return temp 
   
   
# vrača elemnt koji je prvi u redu
    def first(self): 
   
        return self.head.data 
   
   
# vrača veličinu reda
    def size(self): 
   
        temp=self.head 
        count=0
        while temp is not None: 
            count=count+1
            temp=temp.next
        return count 
       
       
# provjera jeli prazan?    
    def isEmpty(self): 
   
        if self.head is None: 
            return True
        else: 
            return False
               
   
# ispis reda
    def printqueue(self): 
           
        print("Elementi u redu su:") 
        temp=self.head 
        while temp is not None: 
            print(temp.data,end="->") 
            temp=temp.next
       
           
# Driver kod   
if __name__=='__main__':  
   
# Prazan red
  queue = Queue() 
   
  print("Operacije nad rtedom implemntiranim sa DDL\n") 
   
# Ubacivanje elemenata u red   
  queue.enqueue(4) 
  queue.enqueue(5) 
  queue.enqueue(6) 
  queue.enqueue(7) 
   
# Ispis reda
  queue.printqueue() 
   
# Ispis prvog elemnta   
  print("\nPrvi elemnt je: ",queue.first()) 
   
# Ispis veličine reda
  print("Veličina reda je: ",queue.size()) 
   
# izbacivanje iz reda
  queue.dequeue() 
  queue.dequeue() 
   
# Ispis nakon izbacivanja
  print("akon dav izbacivanja rede je: ") 
  queue.printqueue() 
   
# Ispsi True ako je prazan
  print("\nRed je prazan (True ili False):",queue.isEmpty()) 
```

## Stog pomoću DLL

Operacije nad Stogom

1. push() : ubacuje novi elemnt na stog i pridružuje pokazivač top.
2. pop() : vrača element na vrhu i top preusmjerava na sljedeći elemnt 
3. top() : samo vrača elemnt na vrhu.
4. size() : Rvrača veličinu  stoga.
5. isEmpty() : provjera d ali je prazan
6. printstack() : ispis svih elemenata.

```python
# Stog implemntiran s DDL
 
  
# Klasa čvor
class Node: 
  
# Function to initialise the node object 
    def __init__(self, data): 
        self.data = data # Assign data 
        self.next = None # Initialize next as null 
        self.prev = None # Initialize prev as null         
          
# Klasa Stack
class Stack: 
    # konstruktor za init liste
    def __init__(self): 
        self.head = None
          
# dodavanje na stog
    def push(self, data): 
  
        if self.head is None: 
            self.head = Node(data) 
        else: 
            new_node = Node(data) 
            self.head.prev = new_node 
            new_node.next = self.head 
            new_node.prev = None
            self.head = new_node 
              
              
# skidanje sa stoga 
    def pop(self): 
  
        if self.head is None: 
            return None
        elif self.head.next is None: 
            temp = self.head.data 
            self.head = None
            return temp 
        else: 
            temp = self.head.data 
            self.head = self.head.next
            self.head.prev = None
            return temp 
  
  
  
  
# dohvačanje elemnta na vrhu
    def top(self): 
  
        return self.head.data 
  
  
# veličina stoga
    def size(self): 
  
        temp = self.head 
        count = 0
        while temp is not None: 
            count = count + 1
            temp = temp.next
        return count 
      
      
# da li je prazan?
    def isEmpty(self): 
  
        if self.head is None: 
           return True
        else: 
           return False
              
  
# Ispis stoga
    def printstack(self): 
          
        print("Elementi stog su:") 
        temp = self.head 
        while temp is not None: 
            print(temp.data, end ="->") 
            temp = temp.next           
          
  
#Driver kod     
if __name__=='__main__':  
  
# Init praznog stoga
  stack = Stack() 
  
# Dodavanje na stog
  print("Stack operacija pomoću Doubly LinkedList") 
  stack.push(4) 
  stack.push(5) 
  stack.push(6) 
  stack.push(7) 
  
# Ispis stoga
  stack.printstack() 
  
# Ispis prvog elementa
  print("\nElement na vrhu je ", stack.top()) 
  
# Ispis veličine stoga
  print("Veličina stoga je ", stack.size()) 
  
# skidanje sa stoga
  stack.pop() 
  stack.pop() 
    
# ispsi nakon skidanja
  stack.printstack() 
    
# ispis True ako je prazan 
  print("\nStog je prazan (True ili False): ", stack.isEmpty()) 
  
```

## Stog implementiran sa jednosruko povezanom listom SSL

Opoeracije nad stogom:

1. push() : ubacuje elemnt na vrh (top)
2. pop() : vrača elemnt na vrhu i izbacije ga sa stoga
3, peek(): vrača elemnt na vrhu.
4. display(): ispis stoga



```python
class Node: 
      
    # Kreiranje čvora u SSL
    def __init__(self,data): 
        self.data = data 
        self.next = None
      
class Stack: 
      
    # glava lsite (Top stoga) podrazumijevano  NULL 
    def __init__(self): 
        self.head = None
      
    # da li je prazan?
    def isempty(self): 
        if self.head == None: 
            return True
        else: 
            return False
      
    # Ubacivanje na stog
    def push(self,data): 
          
        if self.head == None: 
            self.head=Node(data) 
              
        else: 
            newnode = Node(data) 
            newnode.next = self.head 
            self.head = newnode 
      
    # Isbacivanje sa stoga
    def pop(self): 
          
        if self.isempty(): 
            return None
              
        else: 
            # Micanje glave i prebacivanje na sljedeći na vrhu
             
            poppednode = self.head 
            self.head = self.head.next
            poppednode.next = None
            return poppednode.data 
      
    # vraća podatak na vrhu
    def peek(self): 
          
        if self.isempty(): 
            return None
              
        else: 
            return self.head.data 
      
    # ispis cijelo stoga   
    def display(self): 
          
        iternode = self.head 
        if self.isempty(): 
            print("stog je prazan") 
          
        else: 
              
            while(iternode != None): 
                  
                print(iternode.data,"->",end = " ") 
                iternode = iternode.next
            return
          
# Driver kod 
MyStack = Stack() 
  
MyStack.push(34)  
MyStack.push(56) 
MyStack.push(67) 
MyStack.push(11) 
MyStack.push(48) 
MyStack.push(90) 

# Ispis svih elementa 
MyStack.display() 
  
# Ispis elementa na vrhu
print("\nNa vrhu stoga je ",MyStack.peek()) 
  
# brisanje sa stoga
MyStack.pop() 
MyStack.pop() 
  
# ispis elemnata
MyStack.display() 
  
# Ispis elementa na vrhu
print("\nNa vrhu stoga je ",MyStack.peek()) 
  
```
