## Jednostruko povezana lista



Izacivanje duplikata iz jednostruko povezane liste

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
 
 
class LinkedList:
    def __init__(self):
        self.head = None
        self.last_node = None
 
    def append(self, data):
        if self.last_node is None:
            self.head = Node(data)
            self.last_node = self.head
        else:
            self.last_node.next = Node(data)
            self.last_node = self.last_node.next
 
    def get_prev_node(self, ref_node):
        current = self.head
        while (current and current.next != ref_node):
            current = current.next
        return current
 
    def remove(self, node):
        prev_node = self.get_prev_node(node)
        if prev_node is None:
            self.head = self.head.next
        else:
            prev_node.next = node.next
 
    def display(self):
        current = self.head
        while current:
            print(current.data, end = ' ')
            current = current.next
    
       
# izbacivanje duplikata
def remove_duplicates(llist):
    current1 = llist.head
    while current1:
        data = current1.data
        current2 = current1.next
        while current2:
            if current2.data == data:
                llist.remove(current2)
            current2 = current2.next
        current1 = current1.next
 
```

```python
# Driver kod za primjer izbacivanje duplikata
if __name__ == '__main__':
    Ll = LinkedList()
    for i in range(1, 20):
        Ll.append(i)
    print('Lista je:')
    print()
    Ll.display()
    for i in range(10, 16):
        Ll.append(i)
    print()
    print('Lista je:')
    Ll.display()
    print()
    remove_duplicates(Ll)
    print('Lista je:')
     
    Ll.display()
    
    
```

### Izbacivanje zadnjeg elementa iz jednostruko povezane liste


```python
class Node:
    # Singly linked node
    def __init__(self, data=None):
        self.data = data
        self.next = None
class singly_linked_list:
    def __init__(self):
        # Createe an empty list
        self.tail = None
        self.head = None
        self.count = 0

    def append_item(self, data):
        #Append items on the list
        node = Node(data)
        if self.head:
            self.head.next = node
            self.head = node
        else:
            self.tail = node
            self.head = node
        self.count += 1
    
    def delete_item(self, data):
        # Delete an item from the list
        current = self.tail
        prev = self.tail
        while current:
            if current.data == data:
                if current == self.tail:
                    self.tail = current.next
                else:
                    prev.next = current.next
                self.count -= 1
                return
            prev = current
            current = current.next
    def iterate_item(self):
        # Iterate the list.
        current_item = self.tail
        while current_item:
            val = current_item.data
            current_item = current_item.next
            yield val

items = singly_linked_list()
items.append_item('PHP')
items.append_item('Python')
items.append_item('C#')
items.append_item('C++')
items.append_item('Java')

print("Originalna lista:")
for val in items.iterate_item():
    print(val)

print("\nLista nakon izbacivanja zadnjeg elementa :")
items.delete_item('Java')
for val in items.iterate_item():
    print(val)

print("\nLista nakon izbacivanja zadnjeg elementa :")
items.delete_item('C++')
for val in items.iterate_item():
    print(val)
```

## Dvostruko povezana lista 

Izbacivanje duplikata iz dvostruko povezane liste

```python
#Čvor  doubly linked list    
class Node:    
    def __init__(self,data):    
        self.data = data;    
        self.previous = None;    
        self.next = None;    
            
class DoubleLinkedList:    
    #head i tail  doubly linked list    
    def __init__(self):    
        self.head = None;    
        self.tail = None;    
            
    #addNode() dodavanje čvora u DDL   
    def addNode(self, data):    
        #Kreiranje novog čvora    
        newNode = Node(data);    
            
        #Dsli je lista prazna    
        if(self.head == None):    
            #oba tail i head pokazuju na newNode    
            self.head = self.tail = newNode;    
            #  prethodni glave pokazuje na None    
            self.head.previous = None;    
            # next od tail pokazuje na None, jer je zadnji u listi  
            self.tail.next = None;    
        else:    
            #newNode se dodaje nakon tail tako da tail's next pokazuje na newNode    
            self.tail.next = newNode   
            #newNode's prethodni će pokazivati na tail 
            newNode.previous = self.tail    
            #newNode postaje novi tail    
            self.tail = newNode    
            #kako je zadnji pokazuje na None    
            self.tail.next = None   
                
    #removeDuplicateNode() izbacuje duplikate iz liste
    def removeDuplicateNode(self):    
            
        #Chprovjera da li je lista prazna
        if(self.head == None):    
            return;    
        else:    
            #Init , current koji pokazuje na glavu
            current = self.head;    
            while(current != None):    
                #index pokazuje na čvor koji je next od current    
                index = current.next    
                while(index != None):    
                    if(current.data == index.data):    
                        #pospremi duplikat u temp   
                        temp = index;    
                        #index's prethdonog čcotra pokazuje na čvor koji je next od index i zato mičemo duplikat čvora
                        index.previous.next = index.next  
                        if(index.next != None):    
                            index.next.previous = index.previous  
                        #brisanje duplikata temp to None    
                        temp = None;    
                    index = index.next   
                current = current.next    
                            
    #display() ispis   
    def display(self):    
        #trenutni pokazuje na head  
        current = self.head;    
        if(self.head == None):    
            print("Lista je prazna");    
            return;    
        while(current != None):    
            #ispsi i inkrementiranje pointera.    
            #print(current.data, end = ''),    
            print(current.data),    
            current = current.next    
        print();   


# Driver kod
dList = DoubleLinkedList();    
#dodavanje u listu   
dList.addNode(1)    
dList.addNode(2)    
dList.addNode(3)    
dList.addNode(2)   
dList.addNode(2)  
dList.addNode(4)   
dList.addNode(5)    
dList.addNode(3)   
     
print("Originalna lista: ")    
dList.display();    
     
#Izbacivanje duplikata    
dList.removeDuplicateNode()    
     
print("Lista nakon izbacivanja duplikata: ")    
dList.display()    

```

### ZADATAK za samostalno rješavanje 

Napisati metodu u DDL koja će izbacivati svakog k.tog elemnta iz liste počevši od repa.
U nasgtavku je rješenje za izbacivanje od glave


```python
def deleteKthNode(head, k): 
      
        # Da li je porazna
        if (head == None): 
            return None

        if (k == 1): 
            freeList(head) 
            return None

        # Init ptr i prev na početku  
         
        ptr = head 
        prev = None

        # Prolazak kroz listu and brisanje svakog k-tog elem
        count = 0
        while (ptr != None): 

            # Inkrement Node brojača
            count = count + 1

            # da li je brojač jednak k 
            # i ako jeste delete current Node 
            if (k == count): 

                # stavi next od current Node na
                #  next od previous Node 
                # delete(prev.next) 
                prev.next = ptr.next

                # brojač na  = 0 da bi dohvatili sljedeći k-ti 
                 
                count = 0

            # obniovi prev ako count nije 0 
            if (count != 0): 
                prev = ptr 

            ptr = prev.next

        return head 
    
def freeList(node): # oslobađanje liste
    while (node != None): 
        next = node.next
        node = next
    return node     
```

## DDL ispis unazad i naprijed


```python
class Node(object):
    # Doubly linked node
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
	
class doubly_linked_list(object):
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append_item(self, data):
        # Append an item 
        new_item = Node(data, None, None)
        if self.head is None:
            self.head = new_item
            self.tail = self.head
        else:
            new_item.prev = self.tail
            self.tail.next = new_item
            self.tail = new_item

        self.count += 1

    def print_forward(self):
        for node in self.iter():
            print(node)

    def print_backward(self):
        current = self.tail
        while current:
            print(current.data)
            current = current.prev
            
    def iter(self):
        # Iterate the list
        current = self.head
        while current:
            item_val = current.data
            current = current.next
            yield item_val

    def reverse_linked_list(self):
        if self.start_node is None:
            print("The list has no element to delete")
            return 
        p = self.start_node
        q = p.nref
        p.nref = None
        p.pref = q
        while q is not None:
            q.pref = q.nref
            q.nref = p
            p = q
            q = q.pref
        self.start_node = p
    
# Driver kod           
items = doubly_linked_list()
items.append_item('PHP')
items.append_item('Python')
items.append_item('C#')
items.append_item('C++')
items.append_item('Java')

print("Ispis DDL unazad:")
items.print_backward()

print("\nIspis DDL unaprijed:")
items.print_forward()
```
